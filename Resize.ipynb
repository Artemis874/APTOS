```python
from glob import glob

files = glob('./test/*')
```

```python
files
```

```python
import os
import cv2 
from tqdm import tqdm
for filename in tqdm(files):
    #print (filename)
    image = cv2.imread(filename)
    res = cv2.resize(image,(512,512),interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('./test2/' + filename.split('/')[-1], res)
```
